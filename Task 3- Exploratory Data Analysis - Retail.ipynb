## Barplots: Categories, Segments, Ship Mode, and Region

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,10));
sns.countplot(df['Category'],ax=axs[0][0])
sns.countplot(df['Segment'],ax=axs[0][1])
sns.countplot(df['Ship Mode'],ax=axs[1][0])
sns.countplot(df['Region'],ax=axs[1][1])
axs[0][0].set_title('Category',fontsize=12)
axs[0][1].set_title('Segment',fontsize=12)
axs[1][0].set_title('Ship Mode',fontsize=12)
axs[1][1].set_title('Region',fontsize=12)
plt.tight_layout()
```

## Find all the Numeric Columns

```python
numeric_cols = df.select_dtypes(include=['int', 'float']).columns.tolist()
df[numeric_cols]
```

## Find all the Categorical Columns 

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
df[categorical_cols]
```

## Various EDA Graphs

```python
for col in numeric_cols:
    plt.figure(figsize=(16, 6))
    plt.subplot(2, 1, 1)
    plt.title("----------   " + col + "   ----------")
    sns.boxplot(data=df, x=col, showmeans=True, color="violet")
    plt.subplot(2, 1, 2)
    sns.histplot(data=df, x=col)
    plt.axvline(df[col].mean(), color="red", linestyle="-")
    plt.axvline(df[col].median(), color="limegreen", linestyle="--")
    plt.show()
```

## Graph of Sub Categories based on Region 

```python
#Graph of Sub-Category based on Region

plt.figure(figsize = (10, 10))
sns.countplot(x= 'Sub-Category', hue = 'Region', data = df)
plt.xticks(rotation=90)
plt.show()
```

## Density Plots

```python
fig, axs = plt.subplots(ncols=2, nrows = 2, figsize = (10,10))
sns.distplot(df['Sales'], color = 'red',  ax = axs[0][0])
sns.distplot(df['Profit'], color = 'purple',  ax = axs[0][1])
sns.distplot(df['Quantity'], color = 'blue',  ax = axs[1][0])
sns.distplot(df['Discount'], color = 'orange',  ax = axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize = 12)
axs[0][1].set_title('Profit Distribution', fontsize = 12)
axs[1][0].set_title('Quantity distribution', fontsize = 12)
axs[1][1].set_title('Discount Distribution', fontsize = 12)
plt.show()
```

## Profit by State

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['orange','blue'],figsize=(15,8))
plt.title('Profit/Loss & Sales Across States')
plt.xlabel('States')
plt.ylabel('Profit/Loss & Sales')
plt.show()
```

## Statewise Profit Analysis

```python
df['State'].value_counts().head(10)
df_state= df.groupby(['State'])[['Sales','Discount','Profit']].mean()
df_state.head(10)
df_state1=df_state.sort_values('Profit')
df_state1[['Profit']].plot(kind = 'bar', figsize = (10,10))
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

## Density Plot that compares Profit vs Sales

```python
plt.figure(figsize=(5,5))
sns.kdeplot(df['Sales'],color='green',label='Sales',shade=True,bw=25)
sns.kdeplot(df['Profit'],color='purple',label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend()
```

## Profit/ Loss of Sub-Categories

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['purple','red'],figsize=(10,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```
